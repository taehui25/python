```python
import base64

class HYPER_Grader:
    def __init__(self):
        self.answers = {
            1: b"MTUwMA==",
            2: b"ODA=",
            3: b"NQ==",
            4: b"MTA3MDA=",
            5: b"Mw==",
            6: b"MTEw"
        }

    def check(self, task_id, user_answer):
        try:
            expected = base64.b64decode(self.answers[task_id]).decode().strip()
            result = str(user_answer).strip()

            if result == expected:
                print(f"Task {task_id}: ✅ 정답입니다!")
            else:
                print(f"Task {task_id}: ❌ 틀렸습니다.")
                print("당신의 출력:", result)
                print("정답:", expected)

        except Exception as e:
            print(f"Task {task_id}: ⚠ 에러 발생 ({e})")

grader = HYPER_Grader()
print("채점 준비 완료!")
```

## 🍔 문제 1

햄버거 가게에서 주문을 받습니다.

세트 메뉴 가격은 8000원입니다.  
손님이 가진 돈이 부족하다면,  
부족한 금액을 result에 저장하세요.

(돈이 충분하면 0을 저장하세요)

---

### 💡 힌트
- 조건문은 `if`를 사용합니다.
- 부족한 금액은 `가격 - 돈`으로 계산할 수 있습니다.

```python
money = 6500

if money < 8000:
    result = 8000 - money
else:
    result = 0

grader.check(1, result)
```

## 🎯 문제 2

게임에서 플레이어가 몬스터를 처치하여 경험치를 얻습니다.

경험치 리스트가 주어졌을 때,  
가장 큰 경험치를 result에 저장하세요.

---

### 💡 힌트
- `for`문을 사용하여 리스트를 순회합니다.
- 현재 값이 result보다 크면 result를 갱신합니다.

```python
exp = [30, 80, 45, 60, 25]

result = 0

for e in exp:
    if e > result:
        result = e

grader.check(2, result)
```

## 🪓 문제 3

광부가 광석을 캐고 있습니다.

while문을 사용하여 광석을 5번 캘 때까지 반복하고,  
총 채굴 횟수를 result에 저장하세요.

---

### 💡 힌트
- `while`문을 사용합니다.
- 반복 조건은 `count < 5` 형태입니다.
- 반복이 끝난 후 count 값을 result에 저장하세요.

```python
count = 0

while count < 5:
    count += 1

result = count

grader.check(3, result)
```

## 🛒 문제 4

쇼핑몰에서 장바구니 가격을 계산합니다.

상품 가격 리스트에서 총 금액을 계산하여 result에 저장하세요.

---

### 💡 힌트
- `for`문을 사용하여 리스트를 순회합니다.
- 값을 계속 더하려면 누적 변수(`result`)를 사용하세요.

```python
prices = [1200, 3000, 2500, 4000]

result = 0

for price in prices:
    result += price

grader.check(4, result)
```

## 🧑‍🎓 문제 5

학생들의 점수가 주어졌습니다.

for문을 사용하여 60점 이상인 학생 수를 result에 저장하세요.

---

### 💡 힌트
- `for`문으로 리스트를 하나씩 확인합니다.
- 조건은 `score >= 60` 형태로 작성할 수 있습니다.
- 조건을 만족하면 `result`를 1 증가시키세요.

```python
scores = [55, 70, 65, 40, 90]

result = 0

for score in scores:
    if score >= 60:
        result += 1

grader.check(5, result)
```

## 🔢 문제 6

1부터 20까지 숫자 중에서  
짝수만 더한 값을 result에 저장하세요.

---

### 💡 힌트
- `range(1, 21)`을 사용하면 1부터 20까지 반복할 수 있습니다.
- 어떤 수가 짝수인지 확인하려면  
  👉 `숫자 % 2 == 0` 을 사용하세요.

```python
result = 0

for i in range(1, 21):
  if i %2 == 0:
    result += i

grader.check(6, result)
```
